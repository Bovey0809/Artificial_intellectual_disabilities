## 题目: 有10%的概率不执行命令, 那么到达目标的概率是多少?

答案: 
1. 如果执行的话, 概率是 \\(0.8^{5} = 0.32768\\)
2. 执行错误并且到达目标的概率是: \\(0.1^4 \times0.8 = 0.00008\\)
3. 相加, 得到最终答案: 0.32776

### 奖励的序列
- inifinite horizons
    - 如果时间是无穷的, 每个状态的策略选择是固定的.
        - 如果可以永生, 那么做什么都没有意义, 最终的结果是不变的, 已经永生了.
    - 如果时间是有限的, 那么每个状态的选择是不固定的.
- utility of sequences
    - [ ] 概念
- \\(U(S_0 S_1 S_2 ... ) = \sum^{\infty}_{t=0}\gamma^t R(s_t)\leq \sum^\infty_{t=0}\gamma^tR(s_t)=\frac{R_{max}}{1-\gamma}, 0< \gamma < 1 \\)
    - 给奖励乘以一个关于时间的指数, 那么奖励会随着时间的增加减少.
    - 作用: 改变“永生”的状态.

### [假设](https://en.wikipedia.org/wiki/Geometric_series)
- 证明\\(\sum^\infty \gamma^n = \frac{1}{1-\gamma}\\)
    - \\(x = (\gamma^0 + \gamma^1+ \dots)\\)
    - \\(\gamma x = (\gamma^1 + \gamma^2+\dots)\\)
    - \\(x - \gamma x = \gamma^0 = 1\\)
    - \\(\therefore \sum^\infty \gamma^n = \frac{1}{1-\gamma}\\)
